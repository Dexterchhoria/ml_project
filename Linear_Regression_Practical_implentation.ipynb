```python
import pandas as pd
import numpy as np

# URL for the dataset
data_url = "http://lib.stat.cmu.edu/datasets/boston"

# Read the raw data, skip initial rows, and create a DataFrame
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)

# The dataset alternates between feature rows and target rows, so we split and merge accordingly
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])

# The target variable is located in the second part of the data (every second row, column 2)
target = raw_df.values[1::2, 2]

# Convert to DataFrame
columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
df = pd.DataFrame(data, columns=columns)

# Add the target column to the dataframe
df['MEDV'] = target

# Display the first few rows of the dataset
df.head()
```

```python
y = df['MEDV']
X = df.drop(columns=['MEDV'])
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
#stanadardizing the dataset
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#cross validation
from sklearn.model_selection import cross_val_score
```

```python
regression=LinearRegression()
regression.fit(X_train,y_train)
```

```python
mse=cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=10)
```

```python
np.mean(mse)
```

```python
#predection
reg_pred= regression.predict(X_test)
```

```python
reg_pred
```

```python
import seaborn as sns
sns.displot(reg_pred-y_test,kind='kde')
#variance is good it is betw +-10
```

```python
from sklearn.metrics import r2_score
score = r2_score(reg_pred,y_test)
```
